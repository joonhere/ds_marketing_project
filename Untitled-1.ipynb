```python
import weakref
import gc # 가비지 컬렉터를 수동으로 실행하기 위한 모듈

# ----------------------------------------------------
# 객체가 메모리에서 해제될 때(RC=0) 메시지를 출력하는 클래스 정의
# ----------------------------------------------------
class Dog:
    def __init__(self, name):
        self.name = name
        self.owner = None 
        print(f"🐶 Dog '{self.name}' 생성됨.")

    def __del__(self):
        # 객체가 메모리에서 해제될 때 호출됨
        print(f"🗑️ Dog '{self.name}' 해제됨. (메모리 해제 성공)")

class Person:
    def __init__(self, name):
        self.name = name
        self.pet = None   
        print(f"👨 Person '{self.name}' 생성됨.")

    def __del__(self):
        # 객체가 메모리에서 해제될 때 호출됨
        print(f"🗑️ Person '{self.name}' 해제됨. (메모리 해제 성공)")

def create_circular_reference_solved_A():
    owner = Person("Jun")
    dog = Dog("Ari")

    # 1) Person -> Dog: 강한 참조
    owner.pet = dog 

    # 2) Dog -> Person: 약한 참조
    # 약한 참조 객체를 별도 변수에 저장하여 del 이후에도 접근 가능하게 합니다.
    weak_ref_to_person = weakref.ref(owner) 
    
    # Dog 객체의 멤버에 약한 참조를 할당 (선택 사항이지만 개념상 필요)
    dog.owner = weak_ref_to_person

    print("\n--- [순환 참조 설정 후 상태] ---")
    # 약한 참조는 호출(call, ())해야 실제 객체를 얻을 수 있습니다.
    print(f"Person의 약한 참조 (현재 생존 여부): {weak_ref_to_person()}") 
    
    # ----------------------------------------------------
    # 2. 외부 참조 제거 및 메모리 해제 확인
    # ----------------------------------------------------
    print("\n--- [외부 참조 해제 시도] ---")
    
    # 이 두 줄이 실행되면, 각 객체에 대한 외부 참조(지역 변수 owner, dog)가 사라집니다.
    # del은 변수 이름 바인딩을 제거합니다.
    del owner 
    del dog
    
    # 파이썬 가비지 컬렉터를 수동으로 실행합니다. (권장)
    gc.collect()
    
    print("\n--- [GC 실행 후 상태] ---")
    # RC가 0이 되어 Person 객체가 해제되었다면, 약한 참조는 None을 반환합니다.
    # 만약 'None'이 출력되고 '🗑️' 메시지가 출력되었다면 순환 참조가 성공적으로 끊긴 것입니다.
    print(f"Person의 약한 참조 (해제 후 상태): {weak_ref_to_person()}")


# 함수 실행
create_circular_reference_solved_A()
```

```python
import weakref

# ----------------------------------------------------
# 객체가 메모리에서 해제될 때(RC=0) 메시지를 출력하는 클래스 정의
# ----------------------------------------------------
class Dog:
    def __init__(self, name):
        self.name = name
        self.owner = None 
        print(f"🐶 Dog '{self.name}' 생성됨.")

    def __del__(self):
        print(f"🗑️ Dog '{self.name}' 해제됨. (메모리 해제 성공)")

class Person:
    def __init__(self, name):
        self.name = name
        self.pet = None   
        print(f"👨 Person '{self.name}' 생성됨.")

    def __del__(self):
        print(f"🗑️ Person '{self.name}' 해제됨. (메모리 해제 성공)")


# 이 함수 내에서 객체를 생성하고 참조를 설정합니다.
def create_circular_reference_solved_B():
    # 이 함수가 끝날 때 owner와 dog 변수는 스코프 밖으로 나가 RC가 감소합니다.
    owner = Person("Jun")
    dog = Dog("Ari")

    # 1) Person -> Dog: 강한 참조
    owner.pet = dog 

    # 2) Dog -> Person: 약한 참조
    dog.owner = weakref.ref(owner) 

    print("\n--- [함수 내부: 순환 참조 설정 완료] ---")
    print(f"Dog의 약한 참조가 가리키는 대상: {dog.owner()}") 
    
    # 함수 종료 시점 (return)에 도달하면, owner와 dog의 지역 변수 바인딩이 해제됩니다.
    # 약한 참조 덕분에 순환 고리가 끊어져 두 객체 모두 RC가 0이 되고,
    # 메모리 해제 메시지(🗑️)가 즉시 출력됩니다.

# 함수 실행
print("--- [함수 호출 시작] ---")
create_circular_reference_solved_B() 
print("--- [함수 호출 종료] ---")

# 이 출력 사이에 🗑️ 메시지 두 개가 반드시 출력되어야 성공입니다.
# 만약 약한 참조가 아닌 강한 참조였다면, 🗑️ 메시지는 출력되지 않았을 것입니다.
```
